### How many times toss winning team has won the match

```python
ipl.columns
```

```python
twmw=[]
for i in range(0,len(ipl['toss_winner'])):
    if ipl['toss_winner'][i]==ipl['winner'][i]:
        twmw.append(1)
    else:
          twmw.append(0)
```

```python
twmw[0:5]
```

```python
ipl['twmw']=twmw
```

```python
ipl
```

```python
ipl['id'][ipl['twmw']==1].count()
```

```python
ipl['id'][ipl['twmw']==0].count()
```

```python
plt.figure(figsize=(5,5))
plt.pie(ipl['twmw'].value_counts(),  labels=ipl['twmw'].value_counts().keys(), autopct='%0.2f%%')
plt.axis('equal')
```
